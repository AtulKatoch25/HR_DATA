# Visualisation Of Dataset

Comparing DailyRate of Employees through Distplot

```python
sns.distplot(df.DailyRate)
```

As there is only Three Departments in these Dataset most of the employees indulge to "sales department".

```python
plt.pie(df.Department.value_counts(),autopct='%1.1f%%',shadow=True)
plt.legend(df.Department)
```

In this barplot it shows that employees who have more working hours have most numbers of time goes with business travel while
the employees have less number of working hours have less number of business travel.

```python
sns.barplot(x=df.BusinessTravel.head(50),y=df.TotalWorkingYears)
```

Distribution of Relationship Satisfaction with Attrition

```python
plt.figure(figsize = (20,8))
sns.countplot(data = df, x = 'RelationshipSatisfaction', hue = 'Attrition', palette = "Paired")
plt.title('Distribution of RelationshipSatisfaction & Attrition', size = 36)
plt.xlabel("RelationshipSatisfaction", size = 20)
plt.ylabel("Frequency", size = 20)
plt.xticks(size = 20);
plt.yticks(size = 16);
```

In this Visualisation it shows that employees who are more veracious at company.

```python
YearsAtCompany = df['YearsAtCompany'].value_counts().sort_index()
fig, ax = plt.subplots(figsize = (18, 8))
ax.bar(YearsAtCompany.index, YearsAtCompany.values,)
plt.title('Total YearsAtCompany', size = 36)
plt.xlabel('Total YearsAtCompany', size = 24)
plt.ylabel('Frequancy', size = 24)
plt.locator_params('x', nbins = 33)
plt.xticks(rotation = 90, size = 12)
plt.yticks(size = 16)
plt.show()
```

```python
plt.figure(figsize = (20,8))
plt.scatter(df.YearsAtCompany, df.YearsInCurrentRole, color = "blue")
plt.title('YearsAtCompany VS. YearsInCurrentRole', size = 36)
plt.xlabel("YearsAtCompanys", size = 24)
plt.ylabel("YearsInCurrentRole", size = 24)
plt.xticks(size = 16);
plt.yticks(size = 16);
plt.show()
```

Work Life Blance VS Daily Rate

```python

cust_color = ['#5cc6e3', '#3c93b3', '#1f6282', '#003653']
work_life_balance_Daily_Rate = df.groupby(['WorkLifeBalance'])['DailyRate'].mean()
fig, ax = plt.subplots(figsize = (10, 8))
ax.bar(['Low', 'Medium', 'High', 'Very High'], work_life_balance_Daily_Rate.values, color = cust_color)
plt.title('Work Life Balance Vs. Daily Rate', size = 20)
plt.xlabel('Work Life Balance', size = 16)
plt.ylabel('Average Daily Rate', size = 16)
plt.locator_params('x', nbins = 8)
plt.xticks(rotation = 90, size = 16)
plt.yticks(size = 16)
plt.show()
```
